```python
#loading dataframe from a csv file
netflix_df = pd.read_csv(r"C:\Users\acer\Desktop\New folder\phon\netflix_data.csv")
```

```python
#Displaying the Top 5 rows of the dataframe
netflix_df.head(5)
```

```python
#Subseting the columns which are movies from the dataframe
netflix_df_movies_only = netflix_df[netflix_df['type'].str.contains('Movie')]
```

```python
#shows the list of columns in the dataframe
netflix_df_movies_only.columns
```

```python
#Selecting Columns to work with
netflix_df_movies_only_col_subset = netflix_df_movies_only.filter( items = ['title', 'country', 'genre', 'release_year', 'duration'])
```

```python
#Showing the top 5 rows of the subset
netflix_df_movies_only_col_subset.head(5)
```

```python
#Assigning a Parameter to the netflix released year column
release_year = netflix_df_movies_only_col_subset['release_year']
```

```python
#Assigning a Parameter to the netflix movie duration column
movie_duration = netflix_df_movies_only_col_subset['duration']
```

```python
#setting up Parameters to plot the movie dataframe
plt.style.use('ggplot')
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(release_year,movie_duration, s = 10)
ax.set_xlabel('Realeased Year')
ax.set_ylabel('Movie Duration')
ax.set_title('Movie Duration By Year')
plt.show
```

```python
# Initialize an empty list called colors
colors = []

# Iterate 
for lab, row in netflix_df_movies_only_col_subset.iterrows() : 
    if row["genre"] == "Children" : 
        colors.append("red")
    elif row["genre"] == "Documentaries" : 
        colors.append("blue")
    elif row["genre"] == "Stand-Up" :
        colors.append("green")
    else:
        colors.append("black")

# Print the first 10 rows of the colors
print(colors[:10])
```

```python
# Create a scatter plot from subset
plt.style.use("ggplot")
fig = plt.figure()

# Create a scatter plot of Duration vs Release Year
plt.scatter(netflix_df_movies_only_col_subset["release_year"],netflix_df_movies_only_col_subset["duration"],c=colors, s=10)

# Add a title and axis label to the plot
plt.xlabel("Release Year")
plt.ylabel("Duration(min)")
plt.title("Movie Duration by Year of Release")

# Show the plot
plt.show()
```
